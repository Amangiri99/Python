#***Functions***

```
#A function which gives the fribonacci number at the said position
def fribonacci(pos):
    a=1
    b=1
    for i in range(pos):
        temp=a+b
        a=b
        b=temp
    return temp
```

```
print(fribonacci(3))
print(fribonacci(8))
print(fribonacci(7))
```

```
#Now I will plot the data ration of the fribonacci ratio
for i in range(2,20):
    ratio=fribonacci(i)/fribonacci(i-1)
    print(i,ratio)
```

```
#A function which gives the fribonacci number at the said position using user given a,b value
def fribonacci_relative(pos,a,b):
    for i in range(pos):
        temp=a+b
        a=b
        b=temp
    return temp
```

```
fribonacci_relative(3,34,55)
```

```
#A function which gives the fribonacci number at the said position without using user given a,b value
def fribonacci_ol(pos,a=1,b=1):
    for i in range(pos):
        temp=a+b
        a=b
        b=temp
    return temp
```

```
print(fribonacci_ol(6))
print(fribonacci_ol(6,1,1))
```

```
#Fribonacci as a recurssive function 
def fribonacci_recurssive(pos,a=1,b=1):
    if pos<=1:
        return pos
    else:
        return fribonacci_recurssive(pos-2)+fribonacci_recurssive(pos-1)
```

```
fribonacci_recurssive(5)
```

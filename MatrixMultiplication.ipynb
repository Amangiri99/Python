###***Matrix Multiplication A^n***
Note-If we want to copy the whole array we need to do

1.   [item for item in A] (For 1 D) 

1.   [item [:] for item in A] (For nD)



```
def square_matrix(A,n):
    if n==1:
        return A
    else:
        T=square_matrix(A,n-1)
        N=mult(A,T)
        return N
```

```
def mult(A,T):
    out=[
         [0,0],
         [0,0]
    ]
    out[0][0]=A[0][0]*T[0][0]+A[0][1]*T[1][0]
    out[0][1]=A[0][0]*T[0][1]+A[0][1]*T[1][1]
    out[1][0]=A[1][0]*T[0][0]+A[1][1]*T[1][0]
    out[1][1]=A[1][0]*T[0][1]+A[1][1]*T[1][1]
    return out
```

```
A=[
   [1,1],
   [1,0]
]
```

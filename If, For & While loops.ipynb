```
#Program for fribonacci series
a=0
b=1
print(a)
print(b)
for i in range(10):
    temp=a+b
    a=b
    b=temp
    print(temp)
```

```
#Factorial of numbers less than 500 using While Loop
a=1
b=1
print(a)
print(b)
i=2
while b<500:
    i=i+1
    temp=a+b
    a=b
    b=temp
    print(temp)
print("Number of iterations is equal to " +str(i)) #or We can write print("iterations", i ,"times")
```

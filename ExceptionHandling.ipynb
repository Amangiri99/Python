BUT BY MISTAKE IF THE USER ENTERS A STRING THAN THERE WILL BE A ERROR SO TO COME OVER THAT WE HAVE EXCEPTION HANDLING (TRY,EXCEPT)

```
def input_single_float(prompt):
    user_input=input(prompt)
    try:
        output=float(user_input)
    except ValueError:
        print("The Output must be a float")
    return output
```

But if we dont set ouput to NONE there will be a mistake again

```
def input_single_float_inverse(prompt):
    user_input=input(prompt)
    output=None
    try:
        output=1/float(user_input)
    except ValueError:
        print("The Output must be a float")
    except ZeroDivisionError:
        print("Don't Enter zero as input")
    return output
```

```
val=input_single_float_inverse("Enter Number")
```
